```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv(r"C:\Users\HP\OneDrive\Documents\Salary prediction\Salary_Data.csv")
print(data.head())
```

```python
print(data.columns)
```

```python
print(data['Education Level'].unique())
print(data['Job Title'].unique())
```

```python
data = pd.get_dummies(data, columns=['Job Title', 'Education Level', 'Gender'], drop_first=True)
print(data)
```

```python
X = data.drop('Salary', axis=1)
y = data['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
new_data = pd.DataFrame({
    'Years of Experience': [5],
    'Job Title': [1],
    'Job Title': [0],
    'Education Level': [1],
    'Education Level': [0]
})
print(model.predict(new_data))
```
